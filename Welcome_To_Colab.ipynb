# Text Preprocessing

```
reviews=[reviewstext.strip() for reviewstext	in reviews.reviewstext] # remove both the leading and the trailing characters
reviews=[reviewstext for reviewstext in reviews if reviewstext] # removes empty strings, because they are considered in Python as False
reviews[0:10]
```

```
# Joining the list into one string/text
reviews_text=' '.join(reviews)
reviews_text
```

```
# Remove Punctuations
no_punc_text=reviews_text.translate(str.maketrans('','',string.punctuation))
no_punc_text
```

```
# Tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```
from nltk import word_tokenize
text_tokens=word_tokenize(no_punc_text)
print(text_tokens[0:50])
```

```
len(text_tokens)
```

```
# Remove stopwords
from nltk.corpus import stopwords
my_stop_words=stopwords.words('english')

sw_list=['I','The','It','A']
my_stop_words.extend(sw_list)

no_stop_tokens=[word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens)
```

```
# Normalize the data
lower_words=[comment.lower() for comment in no_stop_tokens]
print(lower_words)
```

```
# Stemming (Optional)
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemmed_tokens=[ps.stem(word) for word in lower_words]
print(stemmed_tokens)
```

```
# Lemmatization
nlp=spacy.load('en_core_web_sm')
doc=nlp(' '.join(lower_words))
print(doc)
```

```
lemmas=[token.lemma_ for token in doc]
print(lemmas)
```

```
clean_reviews=' '.join(lemmas)
clean_reviews
```
